### Paths

```python
base_model = "Qwen/Qwen2.5-1.5B"
```

```python
checkpoint_path = "./output_alpha16_r16_cvrp_gamma_train_embed_tok_False_seq7000_b1_ep1/checkpoint-31250/"
```

## Transformers Library

```python
tokenizer_transformers = AutoTokenizer.from_pretrained(base_model)
```

```python
model_transformers = AutoModelForCausalLM.from_pretrained(
    base_model,
    torch_dtype="auto",
    device_map="auto"
)
```

```python
model_transformers = PeftModel.from_pretrained(
    model_transformers,
    checkpoint_path
)
```

```python
model_transformers = model_transformers.merge_and_unload()
```

## Unsloth Library

```python
model_unsloth, tokenizer_unsloth = FastLanguageModel.from_pretrained(
    model_name = base_model,
    max_seq_length = 7000,
    load_in_4bit = False,
)
```

```python
model_unsloth = PeftModel.from_pretrained(
    model_unsloth,
    checkpoint_path,
)
```

```python
model_unsloth = model_unsloth.merge_and_unload()
```

### Check both models compatibilities

```python
tokenizer_transformers.encode("### Response:") == tokenizer_unsloth.encode("### Response:")
```

```python
print(tokenizer_transformers.eos_token == tokenizer_unsloth.eos_token)
print(tokenizer_transformers.eos_token_id == tokenizer_unsloth.eos_token_id)
print(model_transformers.config.eos_token_id == model_unsloth.config.eos_token_id)
```

```python
counter_transformers = 0

for name, param in model_transformers.named_parameters():
    print(name[17:])
    counter_transformers += 1
```

```python
counter_unsloth = 0

for name, param in model_unsloth.named_parameters():
    print(name)
    counter_unsloth += 1
```

```python
print(f"""
Transformers layer count: {counter_transformers}
Unsloth layer count: {counter_unsloth}""")
```

```python
t = model_transformers.model.layers[0].self_attn.q_proj.weight
```

```python
u = model_unsloth.model.layers[0].self_attn.q_proj.weight
```

```python
import torch

print(torch.allclose(
    t.cpu(),
    u.cpu(),
    atol=1e-5
))
```

```python
diff = (t.cpu() - u.cpu()).abs()

print("max:", diff.max().item())
print("mean:", diff.mean().item())
```
